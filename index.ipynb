# Numerical Acoustics

This are some [jupyter](https://jupyter.org/) notebooks that support parts of the course on numerica acoustics at TU Berlin. To use them, clone or copy the repository including all files to your computer. You will need a [Python 3](https://python.org) installation such as [Anaconda](https://anaconda.com/distribution) with the usual scientific packages present. For correct rendering the use of jupyter LaTex [extensions](https://anaconda.org/conda-forge/jupyter_contrib_nbextensions) is recommended.
